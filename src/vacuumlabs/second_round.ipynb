```python
import pandas as pd
import numpy as np
```

```python
with open("data_for_applicants.txt", "r") as f:
    x = f.readlines()

x = [x[i].split("\t") for i in range(len(x))]
df = pd.DataFrame(x, columns=["payment_date", "acc_to", "amount_czk", "acc_from"])

df[["acc_from", "amount_czk", "acc_to"]] = df[["acc_from", "amount_czk", "acc_to"]].astype(int)
df["payment_date"] = pd.to_datetime(df["payment_date"])
df.head()

```
